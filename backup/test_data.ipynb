```python
import csv 
path = "data"
data_versions = ["v1_deontology", 
                 "v2_commonsense", 
                 "v3_moralitylow",
                 "v4_moralityhigh",
                 "v5_conflict"
                 ]
```

```python
import pandas as pd
csv_files = [f"{path}/{version}.csv" for version in data_versions]

def read_and_merge_csv(path, data_versions):
    # Create list of dataframes
    dfs = {}
    
    for version in data_versions:
        file_path = f"{path}/{version}.csv"
        try:
            df = pd.read_csv(file_path)
            dfs[version] = df
        except Exception as e:
            print(f"Error reading {file_path}: {e}")

    return dfs

# merge all data
data = read_and_merge_csv(path, data_versions)
```
